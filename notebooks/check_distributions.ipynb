```python
%matplotlib inline
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

print("Train data age distribution")
root_path = Path("../data/processed_data")
train_csv_path = root_path.joinpath("train.csv")
train_df = pd.read_csv(train_csv_path)
_ = plt.hist(train_df["age"], range(100))
plt.show()
print("Validation data age distribution")
valid_csv_path = root_path.joinpath("valid.csv")
valid_df = pd.read_csv(valid_csv_path)
_ = plt.hist(valid_df["age"], range(100))
plt.show()
print("Test data age distribution")
test_csv_path = root_path.joinpath("test.csv")
test_df = pd.read_csv(test_csv_path)
_ = plt.hist(test_df["age"], range(100))
plt.show()
```
